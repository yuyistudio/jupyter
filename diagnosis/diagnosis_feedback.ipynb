```python

def load_data(input_filename):
    X_res = []
    y_res = []
    rows = []
    c1 = 0
    c0 = 0
    for line in open(input_filename):
        f = line.strip().split()
        X_res.append(float(f[0]))
        y_res.append(float(f[1])/1e5)
        rows.append(f)
    return np.array(X_res, dtype='float64'), np.array(y_res, dtype='float64'), rows

```

```python
import os
files = []
for item in os.walk('./data'):
    files = item[2]
files = files[:4]
idx = 1
for filename in files:
    X, y, rows = load_data('./data/%s' % filename)
    plt.subplot(220 + idx)
    idx += 1
    plt.title(filename)
    plt.xlabel("budget")
    plt.ylabel("predict roi")
    plt.plot(X, y)
    plt.xlim((0, max(X)))
    plt.ylim((0, max(y) * 1.1))
plt.show()
```
