```python
# 下面的import写法是约定俗成的
import matplotlib as mp
# matplotlib本身是面向对象的，为了方便包装成了函数式的调用方式，放到 pyplot 包里面
import matplotlib.pyplot as plt 

# 构造一个二维数组。大部分时候我们的数据都是二维数组
X = np.array([
        [5,3,7],
        [7,9,5],
        [3,5,7],
              ]) 
print '第一个', X[:,0] # 所有这表示所有行，第一列。对照一下结果看是否是这样吧！
print '第二个', X[:,1] # 这表示所有行，第二列
print '第三个', X[0,:] # 这表示第一行，所有列

Y = X[:,0] # 用大写的 Y 表示一个数组，或者叫向量，Vector
plt.plot(Y) # 使用 plot 绘制折线图。现在画一张图，我们指定 Y 的值就行了，横坐标的值先不关心
plt.show() # 绘制完成之后调用 show 才能显示出来。现在验证一下绘制是否正确

```

```python
# 现在同时指定 X 和 Y 进行绘制
plt.plot([1,2,3],[4,5,4]) # 绘制了 (1,4) (2,5) (3,4) 这三个点
plt.show()
```

```python
# 当然也可以这样
plt.plot([1,2,1],[1,2,3])
plt.show()
```

```python
# 更复杂一些
plt.plot(X[:,1], Y)
plt.show()
```

```python
# 那我们如何知道该怎么调用这个函数呢？如何传递其他参数呢？
# 我们可以查阅 pyplot 的文档：http://matplotlib.org/api/pyplot_summary.html
# 找到 plot 这个函数，并点进去查看参数介绍
# 根据文档我们可以这样
plt.plot(X[:,1], Y, 'or') # 绘制的样式变成圆圈，颜色为红色（red）
plt.plot(X[:,0], X[:,2], 'xg') # 绘制的样式变成叉叉，颜色为绿色（green）
plt.xlim(0, 10) # 指定横轴和纵轴的范围
plt.ylim(0, 10)
plt.show()
```

```python
# 甚至可以这样
plt.plot(X[:,1], Y, 'or', X[:,0], X[:,2], 'xg')
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.show()
```

```python
# 下面介绍复杂一些的概念
# figure 表示一张图。上面绘制的都是在默认 figure 上进行的
# subplot 表示图中的一个绘图区域
# 有了这两个概念就可以同时绘制多个区域了

f0 = plt.figure(0) # 切换到第一张图进行绘制
for i in range(6):
    s = plt.subplot(231 + i) # 指定f0图为2行3列（总共6个绘图区域），并获取其中的第 i 个绘图区域
    #s.title('subplot_%s' % i)
    s.axis([1,3,0,6])
    plt.title('figure_%d' % i)
    s.plot([1,2,3],[1,i + 1,1]) # 中间的点的 y 值为 i

plt.figure(1) # 切换到第二张图进行绘制
plt.title('figure1')
plt.plot([1,2,3],[2,1,2])
plt.axis([1,3,1,2])

plt.show() # 绘制完成后进行显示
```
